Pre-Processing

```python
#Seperate X and Y
X=input.drop('Attrition',axis=1)
Y=input['Attrition']
```

```python
X.shape,Y.shape
```

```python
#Train Test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=7)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
#Percentage Of missing value
input.isnull().sum()/input.shape[0]*100
```

```python
#Check column with single value
X_train.nunique()
X_train.drop(['Over18','EmployeeCount','StandardHours'],axis=1,inplace=True)
X_test.drop(['Over18','EmployeeCount','StandardHours'],axis=1,inplace=True)
```

```python
X_train.shape,X_test.shape
```

```python
#Remove Repeated Rows
#
X_train.drop_duplicates(inplace=True)
X_train.shape
```

```python
#Encoding String Column
#Ordinal->label encoding
#Nominal->Hot encoding
# ord_columns=['BusinessTravel']
# X_train['BusinessTravel'].unique()

label_map={'BusinessTravel':{'Travel_Rarely':1,
                             'Non-Travel':0,
                             'Travel_Frequently':2}}
X_train.replace(label_map,inplace=True)
X_train['BusinessTravel'].unique()
X_test.replace(label_map,inplace=True)
X_test['BusinessTravel'].unique()
```

```python
#One Hot encoding
# X_train.dtypes
X_train_ohe=pd.get_dummies(X_train)
X_train.shape,X_train_ohe.shape
# X_train_ohe.columns
```

```python
X_train_ohe.info()
```

```python
#Fill Missing Vales
num_cols=X_train.select_dtypes(exclude='object').columns
num_cols
training_median_value=X_train.loc[:,num_cols].median()
X_train.fillna(training_median_value,inplace=True)
X_test.fillna(training_median_value,inplace=True)
```
